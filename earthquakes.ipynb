# Earthquakes map and classification

```python
###### ------------------- Main libraries -------------------################
import pandas as pd
import numpy as np 
from pprint import pprint
import folium 
from folium.plugins import HeatMap
from folium.plugins import MarkerCluster
import datetime
```

```python
earthquake = pd.read_csv('./earthquakes.csv')
earthquake.head()
```

```python
earthquake['date_time']  = earthquake.Date +' '+ earthquake.Time
earthquake.head(2)
```

```python
earthquake['color'] = np.where(earthquake.Magnitude<=4,'green',
                               np.where(earthquake.Magnitude>=6,'red',
                                       np.where((earthquake.Magnitude<6) & (earthquake.Magnitude>4),'blue','darkblue')))
earthquake.head(2)
```
