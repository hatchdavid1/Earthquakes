# Earthquakes map and classification

```python
###### ------------------- Main libraries -------------------################
import pandas as pd
import numpy as np 
from pprint import pprint
import folium 
from folium.plugins import HeatMap
from folium.plugins import MarkerCluster
import datetime
```

```python
earthquake = pd.read_csv('./earthquakes.csv')
earthquake.head()
```

```python
earthquake['date_time']  = earthquake.Date +' '+ earthquake.Time
earthquake.head(2)
```

```python
earthquake['color'] = np.where(earthquake.Magnitude<=4,'green',
                               np.where(earthquake.Magnitude>=6,'red',
                                       np.where((earthquake.Magnitude<6) & (earthquake.Magnitude>4),'blue','darkblue')))
earthquake.head(2)
```

```python
map1 = folium.Map(location = [19.4978,-99.1269], zoom_start = 10)
locations = earthquake[['Latitude', 'Longitude']]
locationlist = locations.values.tolist()
#for point in range(0,len(locationlist)):
#    folium.Marker(locationlist[point], popup=[general_ags['MUNICIPIO'],general_ags.DEFUNCION]).add_to(map1)
#map1
marker_cluster = MarkerCluster().add_to(map1)
iconname = 'bolt'
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=earthquake['Magnitude'][point],
                 icon=folium.Icon(color=earthquake["color"][point], icon_color='white', icon='glyphicon-pushpin', angle=0, prefix='glyphicon')).add_to(marker_cluster)
#map1.save('./mx_covid.html')
#map1
```
